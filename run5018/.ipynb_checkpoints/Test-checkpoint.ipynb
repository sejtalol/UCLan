## test potential, E<sub>J</sub>

```python
import mymodules as mod
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
nn = 300

filename=('./data/ptcls_info_t%3.3d.dat')%(nn)
x,y,vx,vy,vz,m,pe=np.loadtxt(filename,usecols=[0,1,2,3,4,5,6],unpack=True)

r = [np.sqrt(x[i]**2+y[i]**2) for i in range(len(x))]
Lz = mod.get_ang(x, y, vx, vy)
Ej = mod.get_ej(x, y, vx, vy, vz, pe, 'bar')
```

```python
font = {'family' : 'Times New Roman',
        'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 20,
       }
# test results
fig = plt.figure(figsize=(16, 10), dpi=72, facecolor="white")
ax1 = plt.subplot(211)
ax1.set_xlabel('$Radius$',fontdict=font)
ax1.set_ylabel('$E_{J}$',fontdict=font)
ax1.scatter(r, Lz, c='k', s=5)
```
