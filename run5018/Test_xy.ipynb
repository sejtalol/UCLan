```python
import mymodules as mod
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
nn = 350
dn = 50
x,y,vx,vy,vz,m,pe = np.loadtxt('./data/ptcls_info_t%3.3d.dat'%(nn),usecols=[0,1,2,3,4,5,6],unpack=True)
x2,y2,vx2,vy2,vz2,m2,pe2 = np.loadtxt('./data/ptcls_info_t%3.3d.dat'%(nn+dn),usecols=[0,1,2,3,4,5,6],unpack=True)
```

```python
Ej = mod.get_ej(x,y,vx,vy,vz,pe,'bar')
fig, ax = plt.subplots(figsize=(14,10))
xmin = -4
xmax = 4
ymin = -4
ymax = 4
mod.xy_plot(x,y,Ej,xmin,xmax,ymin,ymax, 100, ax)
ax.set_xlabel('x',fontsize=18)
ax.set_ylabel('y',fontsize=18)
ax.tick_params(direction='out',labelsize=15)
ax.set_title('$E_{J}$, $\Omega_{P}$ = bar',fontsize=20)

mod.plot_r('bar', ax)
```
